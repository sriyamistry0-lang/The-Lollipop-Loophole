```python
from Bio import SeqIO
import pandas as pd

# CRISPR SUGGESTION: STEP 5
# Path to the nucleotide file generated by Prokka
ffn_path = f"{output_dir}/{project_id}.ffn"

def get_crispr_suggestions(report_csv, ffn_file):
    report_df = pd.read_csv(report_csv)
    # Extract the locus tags of the markers we identified earlier
    target_locus_tags = report_df['locus_tag'].tolist()

    crispr_results = []

    # Parse the nucleotide fasta
    for record in SeqIO.parse(ffn_file, "fasta"):
        # Check if this sequence belongs to one of our identified targets
        # Prokka headers usually look like: locus_tag [gene] [product]
        locus_tag = record.id

        if locus_tag in target_locus_tags:
            sequence = str(record.seq)
            gene_name = record.description.split(" ", 1)[1] if " " in record.description else "Unknown"

            # Scan for SpCas9 PAM (NGG)
            # We look for 'GG' and take the 20bp before it
            for i in range(20, len(sequence) - 3):
                if sequence[i+1:i+3] == "GG":
                    grna = sequence[i-20:i]

                    # Logic: We prefer edits in the first 25% of the gene for total knockout
                    pos_pct = (i / len(sequence)) * 100

                    crispr_results.append({
                        'Locus_Tag': locus_tag,
                        'Product': gene_name,
                        'gRNA_Target': grna,
                        'PAM': sequence[i:i+3],
                        'Efficiency_Score': "High (Early)" if pos_pct < 25 else "Moderate",
                        'Position_Pct': round(pos_pct, 2)
                    })

    return pd.DataFrame(crispr_results)

# Run the suggester
if os.path.exists("Final_Pneumo_Report.csv"):
    suggestions_df = get_crispr_suggestions("Final_Pneumo_Report.csv", ffn_path)
    print("\n--- SUGGESTED CRISPR EDITS FOR AMR MARKERS ---")
    # Display the best 10 targets (closest to the start of the gene)
    print(suggestions_df.sort_values(by='Position_Pct').head(10))
    suggestions_df.to_csv("CRISPR_Target_Suggestions.csv", index=False)
```

```python

from Bio import SeqIO
from Bio.Seq import Seq

def interactive_editor(fasta_file):
    # Use global to make these accessible to the Active Site Check cell
    global orig_prot, mut_prot

    tag = input("Enter Locus Tag (e.g., KFONEOIH_00320): ").strip().split(' ')[0]
    print("\n[1] Deletion | [2] Insertion | [3] Substitution")
    m_choice = input("Enter choice (1-3): ").strip()

    for record in SeqIO.parse(fasta_file, "fasta"):
        if record.id == tag:
            dna = str(record.seq)
            orig_prot = str(record.seq.translate(to_stop=True)) # Store globally
            pos = 40

            if m_choice == '1':
                mut_dna = dna[:pos] + dna[pos+1:]
            elif m_choice == '2':
                mut_dna = dna[:pos] + "G" + dna[pos:]
            else:
                mut_dna = dna[:pos] + ("T" if dna[pos] != "T" else "A") + dna[pos+1:]

            mut_prot = str(Seq(mut_dna).translate(to_stop=True)) # Store globally

            print(f"\n--- Result for {tag} ---")
            print(f"Original Length: {len(orig_prot)} aa")
            print(f"Mutated Length:  {len(mut_prot)} aa")
            return

    print(f"Error: '{tag}' not found.")

interactive_editor(ffn_path)
```

```python
def interactive_active_site_check(original, mutated):
    print("--- CUSTOM ACTIVE SITE ANALYSIS ---")
    print("Example Motifs for PBPs: STMK, SSN, KSG")
    print("Example for Efflux Pumps: GXXXG, DXXKN")

    # User inputs motifs of interest separated by commas
    user_input = input("\nEnter motifs to check (e.g., STMK, SSN, KSG): ")
    motifs = [m.strip().upper() for m in user_input.split(',')]

    results = []
    print(f"\nScanning sequences for {len(motifs)} motifs...")

    for motif in motifs:
        # Check if motif exists in the original sequence
        if motif in original:
            # Check if it survived the mutation
            is_present = motif in mutated
            status = "✅ INTACT" if is_present else "❌ DELETED/DISRUPTED"
            results.append({"Motif": motif, "Status": status, "Found_in_Original": "Yes"})
        else:
            results.append({"Motif": motif, "Status": "N/A", "Found_in_Original": "No (Check spelling)"})

    # Display results
    import pandas as pd
    res_df = pd.DataFrame(results)
    print("\n")
    print(res_df)

    # Final Verdict
    deleted_count = len(res_df[res_df['Status'] == "❌ DELETED/DISRUPTED"])
    if deleted_count > 0:
        print(f"\nVERDICT: Knockout successful. {deleted_count} critical motif(s) destroyed.")
    else:
        print("\nVERDICT: Protein structure preserved. Binding pocket may still be functional.")

# Run the check
try:
    interactive_active_site_check(orig_prot, mut_prot)
except NameError:
    print("Error: Please run the Interactive Editor cell first to define your sequences!")
```
